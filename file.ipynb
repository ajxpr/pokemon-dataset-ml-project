#### Visualization

```python
df_top_50 = df.head(50)
```

```python
df_top_50.plot(x = 'Name' , y = 'Total', kind = 'bar' , figsize = (20,10))
plt.xlabel("Name of the Pokemon")
plt.ylabel("Power Level of the Pokemon")
plt.title("Pokemons and their power levels")
```

```python
df_top_50.plot.scatter(x = 'Name' , y = 'Total', figsize = (20,10), s = 100)
plt.xlabel("Name of the Pokemon")
plt.ylabel("Power Level of the Pokemon")
plt.title("Pokemons and their power levels")
```

```python
df_top_50.plot(x = 'Name' , y = 'Total', kind = 'barh' , figsize = (20,10))
plt.xlabel("Name of the Pokemon")
plt.ylabel("Power Level of the Pokemon")
plt.title("Pokemons and their power levels")
```

```python
df_top_50.plot(x = 'Name' , y = 'Total', kind = 'area' , figsize = (20,10))
plt.xlabel("Name of the Pokemon")
plt.ylabel("Power Level of the Pokemon")
plt.title("Pokemons and their power levels")
```
